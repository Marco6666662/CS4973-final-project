## Skew

```python
majority = []
minority = []
k_range = range(30, 500, 30)

for K in k_range:
    # the total population distribution
    #overall_dist = df.minority.value_counts(normalize=True)
    overall_dist = pd.Series([0.9111, 0.0889], name="minority")

    # sample top k distribution
    ranking_dist = df[:K].minority.value_counts(normalize=True)
    
    # calculate the skew
    skew = ranking_dist / overall_dist

    majority.append(skew[0])
    minority.append(skew[1])
    
sns.set(font_scale=1.5)
sns.set(style="whitegrid")  
    
plt.bar(k_range, majority, label='Han', color='orange', alpha=0.7, width=5)
plt.bar(k_range, minority, label='Minority', color='steelblue', alpha=0.7, width=5)

plt.xticks(k_range)
plt.tight_layout() 
plt.legend()
plt.xlabel("choice of K")
plt.ylabel("skew")
plt.title('change of skew based on top K')
```

```python
import numpy as np

def attention(idx, p):
    # idx is rank - 1
    return 100 * (1-p)**idx*p


def get_abr(df, p):
    
    major_df = df[df.minority == 0]
    minor_df = df[df.minority == 1]
    
    major_att= np.mean([attention(idx, p) for idx, row in major_df.iterrows()])
    minor_att= np.mean([attention(idx, p) for idx, row in minor_df.iterrows()])
    
    ABR = min(major_att, minor_att) / max(major_att, minor_att)
    
    return ABR, major_att, minor_att
        
```

```python
sns.set(font_scale=1.5)
sns.set(style="whitegrid")  

k_lst = range(10, 110, 10)
p_lst = [0.0001, 0.01, 0.015, 0.02, 0.05]

i = 1

for k in k_lst:
    
    sub_df = df[:k]
    
    abr_lst = []
    major_att_lst = []
    minor_att_lst = []
    
    for p in p_lst:
        abr, ma_at, mi_at = get_abr(sub_df, p)
        abr_lst.append(abr)
        major_att_lst.append(ma_at)
        minor_att_lst.append(mi_at)
        
    plt.subplot(4, 5, i)
    plt.plot(p_lst, abr_lst, label='abr', color='yellow', alpha=0.7)
    plt.plot(p_lst, major_att_lst, label='Han', color='orange', alpha=0.7)
    plt.plot(p_lst, minor_att_lst, label='Minority', color='steelblue', alpha=0.7)
    plt.gcf().set_size_inches(25, 10)  # Adjust to your desired width and height in inches

    plt.xticks(p_lst)
    plt.legend()
    plt.xlabel("choice of p")
    plt.ylabel("average attention across group")
    plt.title(f'With top k: {k}')

    i += 1
    
plt.tight_layout() 
plt.subplots_adjust(top=0.85)
plt.suptitle('change of average attention based on different choice of p and k')
```

## NDCG

```python
# normalize the six attributes

df["likes"] = (df["likes"] - min(df["likes"])) / (max(df["likes"] - min(df["likes"])))
df["comments"] = (df["comments"] - min(df["comments"])) / (max(df["comments"] - min(df["comments"])))
df["collect"] = (df["collect"] - min(df["collect"])) / (max(df["collect"] - min(df["collect"])))
df["hot_score"] = (df["hot_score"] - min(df["hot_score"])) / (max(df["hot_score"] - min(df["hot_score"])))
df["collection_score"] = (df["collection_score"] - min(df["collection_score"])) / (max(df["collection_score"] - min(df["collection_score"])))
df["related_score"] = (df["related_score"] - min(df["related_score"])) / (max(df["related_score"] - min(df["related_score"])))
```

```python
df['rel_score'] = (df["likes"] + df["comments"] + df["collect"] + df["hot_score"] + df["collection_score"] + df["related_score"]) / 6
```

```python
df.head()
```

```python
ranked_videos = df.sort_values(by='rel_score', ascending=False)
ranked_videos.head(5)
```

```python
from math import log2

def dcg(i, rel):
    return (2**(rel) - 1)/(log2(i+1))
```

```python
dcg_lst = []

for idx, row in df.iterrows():
    df_sub = df[:idx]
    
    dcg_score = np.sum([dcg(i+1, r['rel_score']) for i, r in df_sub.iterrows()])
    dcg_lst.append(dcg_score)
```

```python
df['dcg_score'] = dcg_lst
df.head(5)
```

```python
dcg_lst = []

for idx, row in ranked_videos.iterrows():
    df_sub = ranked_videos[:idx]
    
    dcg_score = np.sum([dcg(i+1, r['rel_score']) for i, r in df_sub.iterrows()])
    dcg_lst.append(dcg_score)
    
ranked_videos['dcg_score'] = dcg_lst
```

```python
ranked_videos.head(5)
```

```python
np.sum(df['dcg_score']) / np.sum(ranked_videos['dcg_score'])
```

```python
k_lst = range(10, 500, 10)
ndcg_lst = []

for k in k_lst:
    ndcg = np.sum(df[:k]['dcg_score']) / np.sum(ranked_videos[:k]['dcg_score'])
    ndcg_lst.append(ndcg)

plt.plot(k_lst, ndcg_lst)
plt.xlabel("choice of k")
plt.ylabel("Normalized Discounted Cumulative Gain")
plt.title("NDCG changes based on top k video")
```

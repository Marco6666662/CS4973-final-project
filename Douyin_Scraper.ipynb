## Then lets turn those meta data to a dataframe!

```python
import pandas as pd

def metadata_to_df(meta_lst):
    
    # Define the columns
    columns = ['author', 'title', 'user_profile_link', 'likes', 'comments', 'collect', 'hot_score', 'collection_score', 'related_score']

    # Create an empty DataFrame with these columns
    df = pd.DataFrame(columns=columns)
    
    for aweme_lst in tqdm(meta_lst, desc="Processing items"):
        
        for aweme in aweme_lst:
            
            if aweme['author'] is None:
                continue
            
            # get the content creater
            author = aweme['author']['nickname']

            # get the video title
            title = aweme['desc']
            
            if 'web_raw_data' not in aweme.keys():
                continue

            # potentially something useful! the recommend_score!
            rec_data = json.loads(aweme['web_raw_data'])['recommend_score']
            hot_score = rec_data['hot_score']
            collection_score = rec_data['collection_score']
            related_score = rec_data['related_score']

            # the statistics of the video!! the "digg_count" is the count of likes
            stats = aweme['statistics']
            collect = stats['collect_count']
            comment = stats['comment_count']
            likes =  stats['digg_count']

            sec_uid = aweme['author']['sec_uid']
            user_profile = f"https://www.douyin.com/user/{sec_uid}"

            new_row = {'author': [author], 'title': [title], 'user_profile_link': [user_profile], 
                       'likes': [likes], 'comments': [comment], 'collect': [collect], 
                       'hot_score': [hot_score], 'collection_score': [collection_score], 'related_score': [related_score]}

            new_row_df = pd.DataFrame(new_row)
            
            df = pd.concat([df, new_row_df], ignore_index=True)

    return df
```
